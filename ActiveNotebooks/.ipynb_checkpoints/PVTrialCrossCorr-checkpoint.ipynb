```python
def crosscorr_mat(trial_mat, inds = np.arange(-29,29)):
    inds = np.arange(-29,30)
    cc_mat = np.zeros([trial_mat.shape[0], trial_mat.shape[0], trial_mat.shape[-1], inds.shape[0]])
    for cell in range(trial_mat.shape[1]):
        for trial0 in range(trial_mat.shape[0]):
            for trial1 in range(trial_mat.shape[0]):

                cc_mat[trial0,trial1,cell,:]= np.correlate(trial_mat[trial0,:,cell],trial_mat[trial1,:,cell],mode='full')
    return cc_mat, inds

# def crosscorr_res(sess):
#     trial_mat = sess.trial_matrices['spks']
# trial_mat[np.isnan(trial_mat)]=1E-5
# fam_trials = (sess.trial_info['LR'] ==-1*sess.novel_arm) & (sess.trial_info['block_number']<5)
# fam_trial_mat = trial_mat[fam_trials,:,:][:,:,sess.fam_place_cell_mask()]

# nov_trials = sess.trial_info['LR'] ==sess.novel_arm
# nov_trial_mat = trial_mat[nov_trials,:,:][:,:,sess.nov_place_cell_mask()]

# fam_cc_mat, inds = crosscorr_mat(fam_trial_mat)
# nov_cc_mat, inds = crosscorr_mat(nov_trial_mat)

# fam_cc_mat = fam_cc_mat.mean(axis=2)
# nov_cc_mat = nov_cc_mat.mean(axis=2)

# fam_cc_com = np.squeeze((fam_cc_mat*inds[np.newaxis,np.newaxis,:]).sum(axis=-1,keepdims=True)/fam_cc_mat.sum(axis=-1,keepdims=True))
# nov_cc_com = np.squeeze((nov_cc_mat*inds[np.newaxis,np.newaxis,:]).sum(axis=-1,keepdims=True)/nov_cc_mat.sum(axis=-1,keepdims=True))

# fam_cc_sym = fam_cc_com*np.tril(-1*np.ones(fam_cc_com.shape),k=-1)
# nov_cc_sym = nov_cc_com*np.tril(-1*np.ones(nov_cc_com.shape),k=-1)
```

```python
trial_mat = sess.trial_matrices['spks']
trial_mat[np.isnan(trial_mat)]=1E-5
fam_trials = (sess.trial_info['LR'] ==-1*sess.novel_arm) & (sess.trial_info['block_number']<5)
fam_trial_mat = trial_mat[fam_trials,:,:][:,:,sess.fam_place_cell_mask()]

nov_trials = sess.trial_info['LR'] ==sess.novel_arm
nov_trial_mat = trial_mat[nov_trials,:,:][:,:,sess.nov_place_cell_mask()]

fam_cc_mat, inds = crosscorr_mat(fam_trial_mat)
nov_cc_mat, inds = crosscorr_mat(nov_trial_mat)

fam_cc_mat = fam_cc_mat.mean(axis=2)
nov_cc_mat = nov_cc_mat.mean(axis=2)

fam_cc_com = np.squeeze((fam_cc_mat*inds[np.newaxis,np.newaxis,:]).sum(axis=-1,keepdims=True)/fam_cc_mat.sum(axis=-1,keepdims=True))
nov_cc_com = np.squeeze((nov_cc_mat*inds[np.newaxis,np.newaxis,:]).sum(axis=-1,keepdims=True)/nov_cc_mat.sum(axis=-1,keepdims=True))

fam_cc_sym = fam_cc_com*(np.tril(-1*np.ones(fam_cc_com.shape),k=-1) + np.triu(np.ones(fam_cc_com.shape,k=0))
nov_cc_sym = nov_cc_com*(np.tril(-1*np.ones(fam_cc_com.shape),k=-1) + np.triu(np.ones(fam_cc_com.shape,k=0))
```
