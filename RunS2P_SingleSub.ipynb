```python
import sys
import os 
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from glob import glob
import h5py

import suite2p as s2p

import s2p_preprocessing as s2p_pp

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
mouse = "4467975.1"
DIRS = ["H:\\2P_Data","I:\\2P_Data"]

sbx_list,savepath_list = [],[]
for DIR in DIRS:
    date_dirs = glob(os.path.join(DIR,mouse,'*'))
#     print(date_dirs)
    for date_dir in date_dirs:
        _,date = os.path.split(date_dir)
        scene_dirs = glob(os.path.join(date_dir,'*'))
        for scene_dir in scene_dirs:
            _,scene_name = os.path.split(scene_dir)
#             print(scene_dirs)
            if scene_name not in ["zstack"]:
                
                data_files = glob(os.path.join(scene_dir, "*.sbx"))
                sbx_list.extend(data_files)
#               
#                 for file in data_files:
#                     sbx_dir,_ = os.path.splitext
#                 sbx_list.extend([os.path.splitext(file)[0] for file in data_files])
                for k,sbx_dir in enumerate(data_files):
                    _,sbx_file = os.path.split(sbx_dir)
                    savepath_list.append(os.path.join("E:",mouse,date,scene_name,sbx_file[:-4]+'.h5'))
                    
                    
                    
                    
print(sbx_list)
# print(savepath_list)
```

```python
h5names = []
for sbx_file,h5_file in zip(sbx_list,savepath_list):
    if os.path.exists(h5_file):
        h5names.append(h5_file)
    else:
        h5names.append(s2p_pp.sbx2h5(sbx_file[:-4],output_name=h5_file,channel_i=-1,force_2chan=True))


```

```python

print(h5names)
```

```python
h5names_trim = [os.path.split(f)[0] for f in h5names]
print(h5names_trim)
```

```python
for _h5name in h5names[3:]:
    print(_h5name)
    try:
        ops = s2p_pp.set_ops(d={'h5py': _h5name,
                                'fast_disk':"F:\\s2p_tmp",
                                'save_path0': os.path.splitext(_h5name)[0],
                                'move_bin':True,
                                'nchannels':2,
                                'tau':.7,
                                'functional_chan':1,
                                'fs':15.46,
                                'roidetect':True,
                                'input_format':"h5",
                                'h5py_key':'data',
                                'nchannels':2,
                                'sparse_mode':True,
                                'allow_overlap':True,
                                'threshold_scaling': 1.25,
                                'aspect': 512./796.})
        ops = s2p.run_s2p(ops=ops)
        binfiles = glob("F:\\s2p_tmp\\*")
        if len(binfiles)>0:
            for binfile in binfiles:
                os.rmdir(binfile)
    except:
        print("%s failed" % _h5name)
```
